```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(-0.5, 0.5, 200)[:, None]
noise = np.random.normal(0, 0.02, x_data.shape)
y_data = np.square(x_data) + noise

x = tf.placeholder(tf.float64,[None, 1])
y = tf.placeholder(tf.float64,[None, 1])

# def middle layer of nn
weights_L1 = tf.Variable(np.random.normal(0, 1, [1,10]))
biases_L1 = tf.Variable(np.zeros((1,10)))
Wx_plus_b_L1 = tf.matmul(x, weights_L1) + biases_L1
L1 = tf.nn.tanh(Wx_plus_b_L1)

# def output layer
weights_L2 = tf.Variable(np.random.normal(0,1,[10,1]))
biases_L2 = tf.Variable(np.zeros((1,1)))
Wx_plus_b_L2 = tf.matmul(L1, weights_L2) + biases_L2
prediction = tf.nn.tanh(Wx_plus_b_L2)

loss = tf.reduce_mean(tf.square(y-prediction))
optimizer = tf.train.AdamOptimizer(0.2).minimize(loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for _ in range(2000):
        sess.run(optimizer, feed_dict={x: x_data, y: y_data})
    
    # get prediction values
    y_pred = sess.run(prediction, feed_dict={x: x_data})
    # plot
    plt.figure()
    plt.scatter(x_data, y_data)
    plt.plot(x_data, y_pred,'k-', lw=5)
    plt.show()
```
